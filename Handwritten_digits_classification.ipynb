# **Importing Packages**

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# **Data loading and preprocessing**

The data is readily available and already preprocessed in the keras library

```python
data = tf.keras.datasets.mnist.load_data()

data
```

```python
(X_train, y_train), (X_test,y_test) = data
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
X_train[0]
```

```python
y_train[:5]
```

```python
#Standardizing X data
X_train_std = (X_train / 255)
X_test_std = (X_test / 255)
```

```python
#Flattening the X data
X_train_flattened = X_train_std.reshape(len(X_train_std), 28*28)
X_test_flattened = X_test_std.reshape(len(X_test_std), 28*28)
```

```python
X_test_flattened.shape
```

```python
#Building, compiling and trining the neural network model
model = Sequential([
    Dense(100, input_shape=(784,), activation='relu'),
    Dense(10, activation='sigmoid')
])

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

model.fit(X_train_flattened, y_train, epochs=5)
```

Model Evalution

```python
model.evaluate(X_test_flattened, y_test)
```

```python
X_test[2]
```

```python
y_pred = model.predict(X_test_flattened)
y_pred[2]
```

```python
np.argmax(y_pred[2])
```

```python
y_pred_labels = [np.argmax(i) for i in y_pred]
y_pred_labels[:5]
```

```python
y_test[:5]
```

```python
cm = tf.math.confusion_matrix(labels=y_test, predictions=y_pred_labels)

plt.figure(figsize=(7,5))
sns.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Prediction')
plt.ylabel('Truth')
```

Numbers in black colored boxes are errors. Remaining colored boxes shown diagonally are successful predictions
