# **Resizing the image using Pillow library**

```python
from PIL import Image
```

```python
img = Image.open('/content/leopard-5873638_1280.jpg')
```

```python
img_resized = img.resize((200,200))
```

```python
img_resized.save('leopard_resized.jpg')
```

```python
#Displaying the image from numpy array

img_res = mpimg.imread('/content/leopard_resized.jpg')
img_plt_res = plt.imshow(img_res)
plt.show()
```

```python
img_res.shape
```

# **Converting RGB images to Grayscale images using OpenCV**

```python
import cv2
```

```python
img = cv2.imread('/content/leopard-5873638_1280.jpg')
```

```python
type(img)
```

```python
img.shape
```

```python
grayscale_img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
type(grayscale_img)
```

```python
grayscale_img.shape
```
